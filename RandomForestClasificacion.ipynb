# Importaci贸n de librerias

```python
###  Importaci贸n de librer铆as

# Librer铆as generales
import pandas as pd
import numpy as np

# Visualizaci贸n
import matplotlib.pyplot as plt
import seaborn as sns

# Modelado y evaluaci贸n
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
```

# Importaci贸n del dataset

```python
from google.colab import drive
drive.mount('/content/drive')

url = '/content/drive/MyDrive/TG_Machine_Learning/datos_datacss.csv'
# Cargar los datos
try:
    data = pd.read_csv(url, sep=",", encoding='latin-1', low_memory=False)
    print("Datos cargados exitosamente!")
except Exception as e:
    print(f"Error al cargar los datos: {e}")
```

# An谩lisis inicial de datos

# An谩lisis de la variable objetivo: Survived

```python
print("Distribuci贸n de la variable objetivo (驴Sobrevive el jugador?):")
print(data['Survived'].value_counts())
sns.countplot(x='Survived', data=data)
plt.title("Distribuci贸n de Supervivencia por Jugador")
plt.show()
```

# Preparaci贸n del dataset de clasificaci贸n

```python
columnas_usadas = ['Survived', 'RoundWinner', 'Team', 'Map', 'RoundStartingEquipmentValue']
df = data[columnas_usadas].dropna().copy()

# Codificaci贸n de variables categ贸ricas
le_team = LabelEncoder()
le_map = LabelEncoder()
le_winner = LabelEncoder()

df['Team'] = le_team.fit_transform(df['Team'])
df['Map'] = le_map.fit_transform(df['Map'])
df['RoundWinner'] = le_winner.fit_transform(df['RoundWinner'])
```

# Definici贸n de variables predictoras y objetivo

```python
y = df['Survived']
X = df.drop('Survived', axis=1)
```

# Divisi贸n del conjunto de datos

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)

print(f"Tama帽o del conjunto de entrenamiento: {X_train.shape}")
print(f"Tama帽o del conjunto de prueba: {X_test.shape}")
```

# Entrenamiento del modelo Random Forest (500 谩rboles)

```python
model_rf = RandomForestClassifier(n_estimators=500,
                                   max_features='sqrt',
                                   class_weight='balanced',
                                   random_state=42,
                                   verbose=2)
model_rf.fit(X_train, y_train)
```

# Evaluaci贸n del modelo

```python
y_pred = model_rf.predict(X_test)
```

# Matriz de confusi贸n

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicci贸n')
plt.ylabel('Valor Real')
plt.title('Matriz de Confusi贸n')
plt.tight_layout()
plt.show()
```

# Reporte de clasificaci贸n

```python
print("\nReporte de Clasificaci贸n:")
print(classification_report(y_test, y_pred))
```

# ROC AUC

```python
roc = roc_auc_score(y_test, y_pred)
print(f"\nROC AUC: {roc:.2f}")
```

# Agregemos una variable m谩s


# Definici贸n de variables para Clasificaci贸n

```python
# Variable objetivo
y = data['Survived']

# Variables predictoras
X = data[['RoundStartingEquipmentValue', 'RoundWinner', 'Map', 'Team']]
```

# Preprocesamiento: Codificar variables categ贸ricas

```python
# Convertimos variables categ贸ricas a variables num茅ricas
X_encoded = pd.get_dummies(X, drop_first=True)
```

# Divisi贸n del conjunto de datos

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.3, random_state=42)

print(f"Entrenamiento: {X_train.shape}")
print(f"Prueba: {X_test.shape}")
```

# Entrenamiento del modelo Random Forest (500 谩rboles)

```python
model = RandomForestClassifier(
    n_estimators=500,
    max_features='sqrt',
    max_depth=10,
    class_weight='balanced',
    random_state=42,
    verbose=1
)

model.fit(X_train, y_train)
```

# Evaluaci贸n del modelo

```python
pred_y = model.predict(X_test)
```

# Resultados del modelo

```python
# Matriz de Confusi贸n
cm = confusion_matrix(y_test, pred_y)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicci贸n")
plt.ylabel("Real")
plt.title("Matriz de Confusi贸n - Clasificaci贸n de Supervivencia")
plt.show()
```

# Reporte de Clasificaci贸n

```python
print("Reporte de Clasificaci贸n:")
print(classification_report(y_test, pred_y))
```

# ROC AUC Score

```python
try:
    roc = roc_auc_score(y_test, pred_y)
    print(f"ROC AUC Score: {roc:.2f}")
except:
    print("No se pudo calcular el ROC AUC Score.")
```

Clase True (Sobrevive):

Menor precisi贸n: hay m谩s falsos positivos (predice que sobrevive, pero no).
